```python
import pandas as pd
import numpy as np
from CNN import CNN
```

```python
train = pd.read_csv("/Users/niklas/Python/AI/Data/digit_train.csv").values
X_train = train[:40000, 1:]
y_train = train[:40000, 0]
X_test = train[40000:, 1:]
y_test = train[40000:, 0]
```

```python
structure = [
    ["conv", 1, 4, 3],
    ["pool", 2],
    ["conv", 4, 8, 3],
    ["FC", 1568, 512, False],
    ["FC", 512, 256, False],
    ["FC", 256, 10, True]
]


model = CNN(structure,
            epochs=15,
            batch_size=100,
            learning_rate=0.0005,
            colour_channels=1,
            picture_h=28,
            picture_w=28,
            adam=True,
            dropout=0.4
            )
model.fit(X_train, y_train)
```

```python
y_pred_train = model.predict(X_train)
print(f"Training accuracy: {round(np.mean(y_pred_train == y_train) * 100, 1)}%")
y_pred_test = model.predict(X_test)
print(f"Test accuracy: {round(np.mean(y_pred_test == y_test) * 100, 1)}%")
```
